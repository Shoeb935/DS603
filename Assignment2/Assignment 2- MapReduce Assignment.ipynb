```python
from collections import defaultdict
import re

#taking date of birth as December 04, 1997
#taking text from book 6 because 12/2=6
#taking pages 4 to 14 from prisoner of Askaban book 6
#function to split text into words
def extract(text):
    allwords = re.findall(r'\b\w+\b', text.lower())
    return allwords

#funtion to update the dictionary
def updatedwordcount(nwords, allwords):
    for word in allwords:
        nwords[word] += 1
#funtion to count the occurance
def nword(file_path):
    nwords = defaultdict(int)
    # Open file with 'utf-8' encoding
    with open(file_path, 'r', encoding='utf-8') as file:
        for text in file:
            allwords = extract(text)
            updatedwordcount(nwords, allwords)
    return nwords
#declaring file path 
path = "C://Users//shoeb//Downloads//File1.txt"
nwords = nword(path)
#printing word and it's count
for w, c in nwords.items():
    print(f'"{w}": {c}')
```

```python
from collections import defaultdict
import re
from spellchecker import SpellChecker

#taking date of birth as December 04, 1997
#taking text from book 6 because 12/2=6
#taking pages 97 to 107 from prisoner of Askaban book 6

#function for extracting not English words in the text
def extract(text, s):
    
    words = re.findall(r'\b\w+\b', text.lower())
    
    notEnglish = [word for word in words if word not in s]
    return notEnglish
#function to update the not english count
def update(nwords, words):
    for word in words:
        nwords[word] += 1
#funtion for word occurances
def notEnglish_num(file_path):
    nwords = defaultdict(int)
    s = SpellChecker()  
   
    with open(file_path, 'r', encoding='utf-8') as file:
        for text in file:
            notEnglish = extract(text, s)
            update(nwords, notEnglish)
    return nwords

path = "C://Users//shoeb//Downloads//File2.txt"
non_english_counts = notEnglish_num(path)
for w, c in non_english_counts.items():
    print(f'"{w}": {c}')
```
